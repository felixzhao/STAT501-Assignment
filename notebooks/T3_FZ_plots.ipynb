```python
plt.figure(figsize=(12, 6))

plt.subplot(1, 2, 1)
plt.scatter(q2_df['Income'], q2_df['Support'])
plt.xlabel("Income")
plt.ylabel("Support")

# Fitting a linear model (trend line)
coefficients = np.polyfit(q2_df['Income'], q2_df['Support'], 1) # Degree 1 for linear
polynomial = np.poly1d(coefficients)
x_range = np.linspace(min(q2_df['Income']), max(q2_df['Income']), 100)
plt.plot(x_range, polynomial(x_range), '-r', label='Trend Line')

plt.legend()
plt.tight_layout()
plt.show()
```

```python
import plotly.graph_objects as go
from sklearn.linear_model import LinearRegression
import pandas as pd
import numpy as np

# Prepare data for linear regression model
X = q4_df[[q4_x_1_col, q4_x_2_col]]
y = q4_df[q4_y_col]
model = LinearRegression().fit(X, y)

# Prepare grid for plotting the plane
x_range = np.linspace(q4_df[q4_x_1_col].min(), q4_df[q4_x_1_col].max(), 10)
y_range = np.linspace(q4_df[q4_x_2_col].min(), q4_df[q4_x_2_col].max(), 10)
xx, yy = np.meshgrid(x_range, y_range)
zz = model.intercept_ + model.coef_[0]*xx + model.coef_[1]*yy

# Create the 3D plot
scatter = go.Scatter3d(x=q4_df[q4_x_1_col], y=q4_df[q4_x_2_col], z=q4_df[q4_y_col], mode='markers')
surface = go.Surface(x=xx, y=yy, z=zz, opacity=0.5)
layout = go.Layout(scene=dict(
                    xaxis=dict(title=q4_x_1_col),
                    yaxis=dict(title=q4_x_2_col),
                    zaxis=dict(title=q4_y_col)))
fig = go.Figure(data=[scatter, surface], layout=layout)

fig.show()
```
