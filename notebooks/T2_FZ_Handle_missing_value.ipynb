```python
df = pd.read_csv(data_path)
print(len(df))
df.head()
```

# NaN row Analysis

```python
int_sup_nan_rows = df[df["Interest"].isna() & df["Support"].isna()]
print(len(int_sup_nan_rows))
int_sup_nan_rows
```

```python
hdi_nan_rows = df[df["Income"].isna() & df["Health"].isna() & df["Edu"].isna() & df["HDI"].isna()]
print(len(hdi_nan_rows))
hdi_nan_rows
```

# Question 2

```python
q2_df = df.dropna(subset=['Income', 'Support'], how='any')
print(len(q2_df))
q2_df.head()
```

```python
q2_df.to_csv(q2_data_path, index=False)
```
